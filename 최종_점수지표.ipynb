# 불법주차 점수 지표 생성

## Minmax Scaling

```python
df2_m = df.copy()
```

```python
# 시설별로 표준화 해줌
from sklearn.preprocessing import minmax_scale
df2_m.iloc[:,1:] = minmax_scale(df2_m.iloc[:,1:])
```

```python
df2_m
```

```python
df2_m_t = df2_m.transpose()
```

```python
df2_m_t = df2_m_t.rename(columns=df2_m_t.iloc[0])
```

```python
df2_m_t = df2_m_t.drop(df2_m_t.index[0])
```

```python
df2_m_t
```

```python
# 행정동 별로 표준화 해줌
from sklearn.preprocessing import minmax_scale
df2_m_t.iloc[:,:] = minmax_scale(df2_m_t.iloc[:,:])
```

```python
df2_m_t
```

```python
df2_m_t = df2_m_t.transpose()
df2_m_t
```

## 불법주차 점수
- 불법주차점수 지표 : 부정 변수가 높을 수록 높게, 긍정 변수가 높을 수록 낮게
- 긍정변수 : 주차장
- 부정변수 : 주차장 외 모든 변수


```python
df2_m_t_w = df2_m_t.copy()
```

```python
df2_m_t_w
```

```python
df2_m_t_w['자동차등록수'] = df2_m_t_w['자동차등록수'] * 0.47
df2_m_t_w['단속cctv'] = df2_m_t_w['단속cctv'] * 0.32
df2_m_t_w['상업시설'] = df2_m_t_w['상업시설'] * 0.047
df2_m_t_w['주거시설'] = df2_m_t_w['주거시설'] * 0.5
df2_m_t_w['교육시설'] = df2_m_t_w['교육시설'] * 0.58
df2_m_t_w['주차장'] = df2_m_t_w['주차장'] * -0.3
df2_m_t_w['불법주정차'] = df2_m_t_w['불법주정차'] * 1
```

```python
df2_m_t_w
```

```python
df2_m_t_w['불법주차점수'] = df2_m_t_w.sum(axis=1)
```

```python
df2_m_t_w
```

```python
df2_m_t_w['불법주차점수'].sort_values(ascending = False)
```

```python
# 10점 척도
df2_m_t_w_10 = df2_m_t_w.copy()

df2_m_t_w_10['불법주차점수'] = df2_m_t_w_10['불법주차점수'] + 1  # 모든 변수 양수 만들기
df2_m_t_w_10['불법주차점수'] = df2_m_t_w_10['불법주차점수'] * 2.85
df2_m_t_w_10['불법주차점수'] = df2_m_t_w_10['불법주차점수'].round(1)
df2_m_t_w_10['불법주차점수'].sort_values(ascending=False)
```
